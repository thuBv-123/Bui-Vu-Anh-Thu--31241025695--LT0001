```python
#exam 6

import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as control


food_type=ctrl.Antecedent(np.arange(1,9,0.1),'food_type')
food_weight=ctrl.Antecedent(np.arange(0,3,0.001),'food_weight')
temperature=ctrl.Antecedent(np.arange(0,60,1),'temperature')
water=ctrl.Antecedent(np.arange(50,1500,1),'water')
cooking_time=ctrl.Consequent(np.arange(15,120,1),'cooking_time')
heat_level=ctrl.Consequent(np.arange(30,100,1),'heat_level')


food_type['soft']=fuzzy.trimf(food_type.universe,[1,1,3])
food_type['medium']=fuzzy.trimf(food_type.universe,[3,5,6])
food_type['hard']=fuzzy.trimf(food_type.universe,[6,9,9])
food_weight['light']=fuzzy.trimf(food_weight.universe,[0,0,1.5])
food_weight['medium']=fuzzy.trimf(food_weight.universe,[0.75,1.5,2.25])
food_weight['heavy']=fuzzy.trimf(food_weight.universe,[1.5,2.25,3])
temperature['cold']=fuzzy.trimf(temperature.universe,[0,0,10])
temperature['warm']=fuzzy.trimf(temperature.universe,[5,25,35])
temperature['hot']=fuzzy.trimf(temperature.universe,[25,42.5,60])
water['low']=fuzzy.trimf(water.universe,[50,400,750])
water['medium']=fuzzy.trimf(water.universe,[400,750,1000])
water['high']=fuzzy.trimf(water.universe,[750,1000,1500])
cooking_time['short']=fuzzy.trimf(cooking_time.universe,[15,30,45])
cooking_time['medium']=fuzzy.trimf(cooking_time.universe,[30,60,90])
cooking_time['long']=fuzzy.trimf(cooking_time.universe,[60,90,120])
heat_level['low']=fuzzy.trimf(heat_level.universe,[30,30,65])
heat_level['medium']=fuzzy.trimf(heat_level.universe,[48,65,82])
heat_level['high']=fuzzy.trimf(heat_level.universe,[65,100,100])



food_type.view()
food_weight.view()
temperature.view()
water.view()
cooking_time.view()
heat_level.view()
```
